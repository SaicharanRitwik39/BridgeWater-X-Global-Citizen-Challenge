```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Data extracted from the image
# Mana (M) and corresponding Dollar ($) values
mana = [500, 2500, 11000, 120000]
dollars = [5, 25, 100, 1000]

# Convert to numpy arrays and reshape for sklearn
X = np.array(mana).reshape(-1, 1)
y = np.array(dollars)

# Fit linear regression model
model = LinearRegression()
model.fit(X, y)

# Predict and plot
x_range = np.linspace(min(mana), max(mana), 500).reshape(-1, 1)
y_pred = model.predict(x_range)

# Plotting
plt.figure(figsize=(8, 5))
plt.scatter(mana, dollars, color='blue', label='Data points')
plt.plot(x_range, y_pred, color='red', label='Regression Line')
plt.xlabel('Mana')
plt.ylabel('Dollars')
plt.title('Linear Regression: Mana vs Dollars')
plt.legend()
plt.grid(True)
plt.tight_layout()

# Regression line formula
slope = model.coef_[0]
intercept = model.intercept_


(slope, intercept)
```
